# Generating a couple of dots

```python
from generator import generate_dots
from plotter import show_plot

GRAPH_SIZE = (500, 500)

dots_num = int(input('Dots num: '))
dots = list(generate_dots(dots_num, *GRAPH_SIZE)) 
dots.sort(key=lambda pair: pair[0]**2 + pair[1]**2)

show_plot(dots)
```

# Linking dots together

```python
from generator import generate_graph

neighbourhood_size = int(input('Neighbourhood size: '))
graph = generate_graph(dots, neighbourhood_size)

show_plot(dots, graph=graph)
```

# Building a dijkstra-tree

```python
from dijkstra import generate_dijkstra_tree

root_id = int(input('Root id: '))
tree = {pair[0]: pair[1]
        for pair in generate_dijkstra_tree(graph, root=root_id)}

show_plot(dots, graph=graph, tree=tree)
```

# Finally, viewing the shortest path

```python
path = {}
path_weight = 0

destination_id = int(input('Destination id: '))
node = destination_id
while node != root_id:
    path_weight += graph[tree[node]][node]
    path[node] = tree[node]
    node = tree[node]

show_plot(dots, graph=graph, path=path)
print('Weight between {} and {} is {}'.format(root_id,
                                              destination_id,
                                              path_weight))
```
